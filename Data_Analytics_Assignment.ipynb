# Data Analytics Assignment - Part 1
Notebook for cleaning and EDA using `cleaned_data.csv`.

```
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('cleaned_data.csv')
print(df.head())
print(df.dtypes)
print(df.shape)

print(pd.DataFrame({'Null Count':df.isnull().sum(),'Null %':df.isnull().mean()*100}))
print('Duplicates:',df.duplicated().sum())
df=df.drop_duplicates()

print(df.describe())
num=df.select_dtypes(include='number')
print('Skewness:')
print(num.skew())

for col in ['Sales','Profit']:
    q1,q3=df[col].quantile([0.25,0.75])
    iqr=q3-q1
    lb=q1-1.5*iqr
    ub=q3+1.5*iqr
    print(col,'Outliers',((df[col]<lb)|(df[col]>ub)).sum())

plt.plot(df.index,df['Sales']); plt.title('Sales Line Plot'); plt.show()
df.groupby('Category')['Sales'].mean().plot.bar(title='Average Sales by Category'); plt.show()
plt.hist(df['Sales'],bins=20); plt.title('Sales Histogram'); plt.show()
plt.scatter(df['Sales'],df['Profit']); plt.title('Sales vs Profit'); plt.show()
df.boxplot(column='Sales',by='Category'); plt.suptitle(''); plt.show()

print('Pearson')
print(num.corr())
print('Spearman')
print(num.corr(method='spearman'))
print(df.groupby('Category')['Sales'].agg(['mean','std','count']))

df.to_csv('cleaned_data.csv',index=False)
```
